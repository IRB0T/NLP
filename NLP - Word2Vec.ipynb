```python
import re
import nltk
from gensim.models import Word2Vec
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
text_data = ""
with open('all_tweets_data.txt',encoding='UTF-8') as f:
    for i in f:
        text_data+=i
```

```python
text = re.sub('[^a-zA-z.]', ' ', text_data)
text = text.lower()
```

### Tokenizing Sentences

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
new_sentences = []
for i in range(len(sentences)):
    if sentences[i] == '.':
        continue
    else:
        new_sentences.append(sentences[i])
        
```

```python
new_sentences
```

```python
new_words = [nltk.word_tokenize(sentence) for sentence in new_sentences]
```

```python
new_words_list=[]
for i in new_words:
    if "." in i:
        i.remove(".")
    new_words_list.append(i)
```

```python
new_words_list
```

```python
for i in range(len(new_words_list)):
    new_words_list[i] = [word for word in new_words_list[i] if word not in stopwords.words('english')]
```

```python
model = Word2Vec(new_words_list,min_count=2)
```

```python
words = model.wv.vocab
```

```python
words
```

```python
similar = model.wv.most_similar('science')
similar
```
