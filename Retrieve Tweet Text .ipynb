```python
import json
import tweepy
```

```python
CONSUMER_API_KEY = ""
CONSUMER_API_SECRET_KEY = ""
ACCESS_TOKEN = ""
ACCESS_TOKEN_SECRET = ""
```

```python
auth = tweepy.OAuthHandler(CONSUMER_API_KEY,CONSUMER_API_SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN,ACCESS_TOKEN_SECRET)
```

```python
api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True, compression=True)
```

```python
user_id = "ProfFeynman"
```

```python
count = 200
user_tweets = api.user_timeline(screen_name=user_id,count=count,tweet_mode="extended")
```

```python
data = []
for tweet in user_tweets:
    data.append(tweet.full_text)
all_tweets_text = ". ".join(data)
```

```python
all_tweets_text
```

```python
import re
```

```python
def change_urls(s):
    new_text = re.sub(
        "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\), ]|(?:%[0-9a-fA-F][0-9a-fA-F]))+",
        " " + " " + " ",s,    
    )
    return new_text
```

```python
new_tweets = change_urls(all_tweets_text)
```

```python
import emoji
```

```python
def strip_emoji(text):
    new_text = re.sub(emoji.get_emoji_regexp(), r"", text)
    return new_text
```

```python
new_tweets = strip_emoji(new_tweets)
```

```python
def strip_newline(text):
    new_text=re.sub("\n","",text)
    return new_text
```

```python
new_tweets = strip_newline(new_tweets)
```
