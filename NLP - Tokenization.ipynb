```python
import nltk
```

```python
nltk.download()
```

```python
text_data = ""
with open('all_tweets_data.txt',encoding='UTF-8') as f:
    for i in f:
        text_data+=i
```

### Tokenizing Sentences

```python
sentences = nltk.sent_tokenize(text_data)
```

```python
sentences
```

### Tokenizing Words

```python
words = nltk.word_tokenize(text_data)
```
