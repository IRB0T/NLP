```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
#nltk.download()
```

```python
text_data = ""
with open('all_tweets_data.txt',encoding='UTF-8') as f:
    for i in f:
        text_data+=i
```

### Tokenizing Sentences

```python
sentences = nltk.sent_tokenize(text_data)
```

```python
sentences
```

```python
new_sentences = []
for i in range(len(sentences)):
    if sentences[i] == '.':
        continue
    else:
        new_sentences.append(sentences[i])
        
```

```python
new_sentences
```

```python
Lemmatizer = WordNetLemmatizer() 
```

```python
corpus = []
for i in range(len(new_sentences)):
    #print(new_sentences[i])
    s = re.sub('[^a-zA-z]', ' ', new_sentences[i])
    #print(s)
    s = s.lower()
    #print(s)
    s = s.split()
    #print(s)
    s = [Lemmatizer.lemmatize(word) for word in s if not word in set(stopwords.words('english'))]
    #print(s)
    s = ' '.join(s)
    #print(s)
    corpus.append(s)
    #break
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
c = TfidfVectorizer()
```

```python
x = c.fit(corpus)
```

```python
x.get_feature_names()
```

```python
x1 = c.transform((corpus))
```

```python
import pandas as pd
pd.DataFrame(x1.toarray(), columns=x.get_feature_names())
```
